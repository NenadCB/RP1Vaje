```python
import matplotlib.pyplot as plt
```

```python
# tortni
fig, ax = plt.subplots(figsize=(4,4))

rezine = [20, 80, 30, 40]
nazivi = ['jagode', 'banane', 'breskve', 'ostalo']
ax.pie(rezine, explode=[0, 0, 0.1, 0.2], labels=nazivi);
```

```python
#stolpčni graf
fig, ax = plt.subplots(figsize=(4,4))

x = [1, 2, 3, 4, 5]
y = [20, 80, 30, 40, 10]

ax.bar(x, y, width=0.9, color='olive')
ax.set_title('stolpični graf')
ax.set_xlabel('meseci')
ax.set_ylabel('količina padavin [mm]')

fig.suptitle('prečudoviti graf')
```

```python
# raztreseni

fig, ax = plt.subplots(figsize=(4,4))

x = [1, 2, 3, 4, 5]
y = [20, 80, 30, 40, 10]
z = [30, 90, 40, 50, 20]

ax.scatter(x, y, s=200, c='orange')
ax.scatter(x, z, s=[200, 150, 400, 300, 100], marker='*')
```

```python
#črtni graf

fig, ax = plt.subplots(figsize=(4,4))

x = [1, 2, 3, 4, 5]
y = [20, 80, 30, 40, 10]
z = [30, 90, 40, 50, 20]

ax.plot(x, y, label='1. letnik', c='red', ls=':')
ax.plot(x, z, label='2. letnik', c='green', ls='-.')
ax.legend()
ax.axis([-1, 10, -10, 50])
ax.grid()
```

```python
#več grafov

fig, ax = plt.subplots(1, 2, figsize=(8,4))

crte, pike = ax

crte.plot(x, y, label='1. letnik', c='red', ls=':')
pike.scatter(x, z, s=[200, 150, 400, 300, 100], marker='*')
```

```python
#več grafov

fig, ax = plt.subplots(2, 2, figsize=(8,4))

ax1 = ax[0,0]
ax2 = ax[0,1]
ax3 = ax[1,0]
ax4 = ax[1,1]

ax1.plot(x, y, label='1. letnik', c='red', ls=':')
ax1.set_title('črtni graf')
ax4.scatter(x, z, s=[200, 150, 400, 300, 100], marker='*')
ax4.set_title('razstreseni')

ax2.pie(rezine, explode=[0, 0, 0.1, 0.2], labels=nazivi);

fig.tight_layout()
fig.suptitle('naslov cele slike')

fig.savefig('primergrafa.png')
```
